# **Data Visualisation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

Two methods: 

1. Basic Matplotlib commands
2. Object Oriented method

#### 1 Basic Commands

```python
plt.plot(x, y, "red")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("Title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")

plt.subplot(1, 2, 2)
plt.plot(y, x, "g*-")
```

#### **2 Object Oriented**

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, right, width, height (0-1)
axes.plot(x, y, "blue")
axes.set_label("x-axis")
axes.set_label("y-axis")
axes.set_title("Title")
```

```python
fig1 = plt.figure()
fig2 = plt.figure()

axes1 = fig1.add_axes([0.1, 0.1, 0.4, 0.4])
axes1.plot(x, y, "red")
axes1.set_xlabel("x-axis 1")
axes1.set_ylabel("y-axis 1")
axes1.set_title("title 1")

axes2 = fig2.add_axes([0.1, 0.1, 0.4, 0.4])
axes2.plot(y, x, "yellow")
axes2.set_xlabel("x-axis 2")
axes2.set_ylabel("y-axis 2")
axes2.set_title("title 2")
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, "r")
axes.set_xlabel("x-axis")
axes.set_ylabel("y-axis")
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
for ax in axes:
    ax.plot(x, y, "b")
    ax.set_xlabel("x-axis")
    ax.set_ylabel("y-axis")
    ax.set_title('title')
```

```python
# to avoid overlapping, use tight_layout()

fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, "b")
    ax.set_xlabel("x-axis")
    ax.set_ylabel("y-axis")
    ax.set_title('title')

fig    
plt.tight_layout()
```

```python
#dpi = dots per inch

fig = plt.figure(figsize=(8,4), dpi=100)

fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x**2, label = "x**2")
axes.plot(y, y**3, label = "x**3")
axes.legend(loc = 0) #matpplotlib decides the optimal position of the legend

# loc = 1 --> upper right corner
# loc = 2 --> upper left corner
# loc = 3 --> lower left corner
# loc = 4 --> lower right corner
```
